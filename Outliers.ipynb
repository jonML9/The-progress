```python
import matplotlib.pyplot as plt
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)
incomes = np.append(incomes, 10000000)
plt.hist(incomes, 100)
plt.show
```

```python
incomes.mean()
```

```python
def throw_outliers(data):
    med = np.median(data)
    st = np.std(data)
    filtered = [e for e in data if (med - 2 * st < e < med + 2 * st)]
    return filtered

filtered = throw_outliers(incomes)

plt.hist(incomes, 100)
plt.show
```

```python
np.mean(filtered)
```
