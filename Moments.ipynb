# Moments: Mean, Variance, Skew, Kurtosis


```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.normal(10, 10, 10000)

plt.hist(values, 50)
plt.show()
```

## Mean

```python
np.mean(values)
```

### Variance

```python
np.var(values)
```

## Skew 
### (since it is well centered around 0, it should be nearly zero

```python
import scipy.stats as sp
sp.skew(values)
```

## Kurtosis
### (shows the sharpness of the data tail. For a normal distribution it should be zero)


```python
sp.kurtosis(values)
```
