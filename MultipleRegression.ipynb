# Multiple Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('http://cdn.sundog-soft.com/Udemy/DataScience/cars.xls')
df.columns
df['Cruise'].unique()
```

## Turning str values into int values

```python
from sklearn.preprocessing import LabelEncoder
encode_model = LabelEncoder()
encode_type = LabelEncoder()
encode_trim = LabelEncoder()
encode_make = LabelEncoder()

df['Model'] = encode_model.fit_transform(df['Model'])
df['Type'] = encode_type.fit_transform(df['Type'])
df['Trim'] = encode_trim.fit_transform(df['Trim'])
df['Make'] = encode_make.fit_transform(df['Make'])
```

## Identifying features and the target value. Then fitting it to the standart scaler so that the model weighs everything 'equally'

```python
from sklearn.preprocessing import StandardScaler

X = df.drop(['Price'], axis=1)
y = df['Price']

scale = StandardScaler()
X_scaled = scale.fit_transform(X)
```

## Fitting the model into a Linear Regression and determing how well the data fits the line

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_scaled, y)

print("R² score:", model.score(X_scaled, y))
```

## Finding out what numbers were given to each of the str values

```python
mapping_make = pd.DataFrame({
    'Make': encode_make.classes_,
    'Code': range(len(encode_make.classes_))
})
print(mapping_make)


mapping_model = pd.DataFrame({
    'Model': encode_model.classes_,
    'Code': range(len(encode_model.classes_))
})
print(mapping_model)


mapping_type = pd.DataFrame({
    'Type': encode_type.classes_,
    'Code': range(len(encode_type.classes_))
})
print(mapping_type)


mapping_trim = pd.DataFrame({
    'Trim': encode_trim.classes_,
    'Code': range(len(encode_trim.classes_))
})
print(mapping_trim)
```
