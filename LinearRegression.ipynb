# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats as st
```

```python
df = pd.read_csv('experience_salary.csv')
df.tail
```

```python
X = df['YearsExperience']
Y = df['Salary']
plt.scatter(x=X, y=Y)
```

## A line of best fit

```python
a, b = np.polyfit(X, Y, 1)
print(a, b)
plt.scatter(X, Y)
plt.plot(X, a*X + b, color='red', linewidth=2)
plt.show()
```

## Using r-squared to see whether how well the data fits the line

```python
slope, intercept, r_value, p_value, std_err = st.linregress(x=X, y=Y)
r_value ** 2
```

## Proving that there is a positive correlation between years of experience and salary

```python
np.corrcoef(x=X, y=Y)
```
